# 1. Introduction and Motivation

This study investigates how the number of children a woman has influences her career outcomes, specifically focusing on wages, promotion opportunities, and labor force participation. The central question we aim to test is whether motherhood significantly affects women's professional advancement and earnings potential.

While structural barriers against working mothers are well-documented in China and many other countries—such as discrimination during pregnancy, reduced access to promotions, and wage penalties due to childcare responsibilities—the extent and mechanisms of these penalties in the U.S. labor market remain underexplored. The United States offers a different cultural, legal, and economic context, making it a valuable setting to examine whether similar patterns exist and, if so, to what degree.

![INTRO](INTRO.jpg)

Understanding the motherhood penalty is crucial for policymakers, employers, and women navigating career planning. If the U.S. has developed successful strategies to mitigate motherhood-related career setbacks, identifying these factors could offer important lessons for shaping more equitable workplace policies globally, including in China. Conversely, if significant penalties persist, it highlights the urgent need for reforms even in advanced economies.

Supporting evidence from earlier research (Kahn et al., 2014; Even and Macpherson, 2001) demonstrates that motherhood penalties are particularly acute in early career stages but tend to diminish over time. However, these findings are based on outdated data and often neglect intersectional dimensions like race and class. More recent reports (University of Chicago and AP-NORC, 2023) reveal shifting societal attitudes toward parenthood and work, yet they fall short of quantifying the economic impacts. By leveraging up-to-date U.S. labor market data and applying an intersectional lens, this study aims to bridge these gaps and provide a more comprehensive understanding of how childbearing affects women’s economic trajectories.

## Factors Influencing Women’s Income: Six Recent Studies

### 1. U.S. Gender Wage Gap Trends and Explanations (Blau & Kahn, 2017)

#### Research Scope & Methods

Economists Francine Blau and Lawrence Kahn conducted a comprehensive literature review coupled with empirical analysis using U.S. data (Panel Study of Income Dynamics, 1980–2010). They aimed to identify how education, experience, occupation, industry, and other demographic factors explain the gender wage gap. Their study tracks the evolution of this gap over three decades and employs statistical decompositions to quantify the impact of each factor on women's earnings relative to men.

#### Key Findings

- **Narrowing Yet Persistent Gap:**
  - The gender wage gap narrowed significantly from 1980 to 2010 but stagnated among high earners. High-earning women still earned substantially less compared to their male counterparts, suggesting a greater disparity at the top of the income distribution.

- **Education and Experience:**
  - Traditional human capital factors (education and experience) now account for very little of the remaining wage gap, as women have largely matched men in these areas.

- **Occupational and Industrial Segregation:**
  - Women continue to predominantly occupy lower-paying industries and occupations. This occupational and industry segregation significantly contributes to the persistent wage gap, more so than education or other credentials.

- **Family Roles and Work Hours:**
  - Career interruptions and fewer work hours, often due to caregiving responsibilities, negatively impact women's earnings, especially in high-skilled, high-paying professions. This phenomenon, known as the "flexibility penalty," disproportionately affects women who prioritize flexible schedules.

- **Discrimination and Unexplained Gap:**
  - Even after accounting for measurable factors like education and occupation, a notable wage gap remains unexplained. Strong evidence from experimental research indicates ongoing discrimination in pay and promotion opportunities.

- **Emerging Factors:**
  - Psychological or non-cognitive skills (e.g., risk tolerance, negotiation skills) have a limited impact on the wage gap compared to occupational and industrial segregation. Such factors modestly influence earnings differences but are not major contributors.

#### Methodological Notes

The study combines quantitative analysis with an extensive literature review. Quantitative data from the U.S. Panel Study of Income Dynamics is analyzed to decompose the wage gap by observable factors. The subsequent literature review contextualizes these findings, incorporating evidence from experimental and international studies. This dual-method approach underscores the robustness of the findings, emphasizing occupational segregation, work-family dynamics, and potential discrimination as significant, ongoing contributors to the gender wage gap.

### 2. Global Analysis of Gender Pay Gaps (ILO Global Wage Report, 2018/19)

#### Research Scope & Methods

The International Labour Organization’s (ILO) 2018/19 Global Wage Report provides a quantitative global analysis of gender pay inequality, utilizing data from over 70 countries covering approximately 80% of wage employees worldwide. It employs various metrics such as average hourly, median, and monthly earnings gaps, and conducts decomposition analyses to identify how worker characteristics and unexplained factors contribute to gender wage disparities.

#### Key Findings

- **Global Gender Wage Gap:**
  - Globally, women earn approximately 16% less per hour on average than men. However, the gap varies significantly by country, ranging from -10% in the Philippines (women earning slightly more than men) to 34% in Pakistan.

- **Unexplained Wage Gap:**
  - In most countries, especially high-income ones, a substantial part of the wage gap remains unexplained after accounting for observable factors like education and job type. This points to discrimination, gender norms, or other unmeasured variables as significant causes.

- **Occupational Segregation:**
  - Women commonly occupy lower-paid jobs or sectors, which significantly contributes to wage disparities. Educational disparities, however, have minimal impact since younger women frequently match or exceed men's education levels globally.

- **Factor-weighted Gender Gap:**
  - A refined metric adjusting for industry and job type increases the global gender wage gap estimate from 16% to about 19%, indicating women earn less even within the same job categories.

- **Wage Distribution Patterns:**
  - In high-income countries, the wage gap widens significantly at higher earnings levels, illustrating a clear "glass ceiling." Conversely, low- and middle-income countries experience larger gaps at lower wage levels, highlighting disparities in informal or minimum-wage employment.

- **Motherhood Penalty:**
  - Women with children often face lower earnings due to career interruptions, part-time employment, and biases, exacerbating gender pay disparities. Insufficient childcare and unequal family responsibilities restrict women’s career growth.

#### Important Notes / Policy Implications

The report emphasizes policy interventions to address both measurable and deeper structural inequalities. Recommended policies include improving equal pay legislation, enhancing pay transparency, raising wages in women-dominated industries, and promoting equitable family responsibilities through improved childcare, parental leave, and flexible working arrangements.

### 3. Education, Race, and Gender Pay Disparities in the U.S. (Budig, Lim & Hodges, 2021)

#### Research Scope & Methods

Sociologists Michelle Budig, Misun Lim, and Melissa Hodges use an intersectional, quantitative approach to disentangle how race and education interact to affect women’s earnings in the United States. They analyze data from the National Longitudinal Survey of Youth 1979–2012, a representative panel following Americans over decades. The study investigates two primary factors explaining wage disparities: (1) educational attainment and field of study—whether minority women attain less education or pursue lower-paying majors—and (2) returns to education—whether women, particularly Black women, receive lower wage returns for equivalent educational credentials compared to other groups. They employ regression decomposition techniques to quantify the portion of wage gaps attributed to educational differences and random-effects models to examine differential wage returns to education.

#### Key Findings

- **Educational Attainment vs. Gender Gaps:**
  - Differences in education account for a significant share of racial pay gaps but do not explain the gender pay gap. Specifically, unequal education levels and college majors explain about 13–23% of the earnings gap between Black and white workers. However, these educational differences explain none of the gender pay gap between women and men. Women have attained sufficient educational parity with men, indicating other factors drive the persistent gender gap. Equalizing educational attainment could significantly reduce racial income disparities but would not eliminate the gender gap.

- **Lower Returns to Education:**
  - The study identifies clear evidence of unequal rewards for the same education. Men generally enjoy higher wage returns to each educational level compared to women, with this discrepancy being most pronounced for Black women. Black women experience particularly lower returns at the master’s degree level, indicating labor market inequalities extending beyond educational access, likely due to discrimination or occupational barriers.

- **Double Disadvantage for Black Women:**
  - Using an intersectional lens, the authors highlight that Black women face compounded disadvantages in earnings. Historically lower educational access (improving in recent generations) combined with lower returns on education means Black women are multiply disadvantaged. Even highly educated Black women earn significantly less than predicted if they were male or of another race with equivalent education. This dual disadvantage—educational attainment gaps and lower returns—substantially contributes to the large income disparity experienced by Black women.

- **Graphical Results & Statistical Illustrations:**
  - The study includes figures or charts from the decomposition analysis. A key visual finding is that equalizing educational attainment between Black and white women notably reduces the racial pay gap but leaves the gender gap largely intact across education levels, highlighting persistent occupational segregation and discrimination. Graphical representations clearly illustrate lower wage gains per additional degree for Black women compared to other groups, underscoring lower returns to education as a significant explanatory factor.

#### Methodological Notes

The study effectively distinguishes between group differences in attributes (education) and differences in how attributes are rewarded (returns to education), showing structural inequities in labor market valuation. Using a longitudinal panel dataset (1979–2012) enhances findings by capturing career-long wage trajectories, confirming the persistence of gender and racial gaps over time. The intersectional approach reveals that improving educational attainment alone is insufficient; policies must also target unequal pay structures. Recommended measures include stronger anti-discrimination enforcement and initiatives to guide minority women into higher-paying fields and roles.

### 4. Motherhood and the Persistent Gender Earnings Gap (Kleven et al. 2019)

#### Research Scope & Methods

 Kleven and colleagues provide a clear statistical analysis of how family structure—particularly parenthood—affects women's earnings. Using administrative data from Denmark, this study tracks career paths of men and women before and after they become parents. They employ an event study design, aligning wage profiles at the birth of the first child and measuring the impact over time. Denmark, known for its generous parental leave and high female workforce participation, provides a critical context for isolating the effect of children on careers. The analysis covers cohorts from 1980 to 2013, examining both immediate and long-term earnings outcomes. The study also explores changes in women's job characteristics (occupation, sector, employer) after motherhood and intergenerational effects.

![Figure: Women’s earnings (pink line) drop significantly after the birth of their first child, while men’s earnings (blue line) remain essentially unchanged. This “child penalty” in a Danish study shows women’s income trajectory diverging sharply downward immediately after childbirth, with a long-run gap of around 20%](LR1.png)

#### Key Findings

- **Large, long-term “child penalty” for women’s earnings**
  - Childbirth creates a significant and enduring divergence in earnings between genders. Women experience a sharp decline in earnings immediately following childbirth, which persists long-term, with women's earnings approximately 20% lower than their childless counterparts even a decade later. Fathers’ earnings remain largely unaffected, highlighting a clear gender disparity attributed to parenthood.

- **Mechanisms—hours, participation, and job choice:**
  - Three primary channels contribute to the motherhood penalty: reduced labor force participation, fewer working hours, and lower hourly wages. After childbirth, many women temporarily exit the workforce or shift to part-time roles, directly lowering earnings. Those who remain employed frequently transition into more flexible but lower-paying jobs, often in the public sector, leading to slower career progression and permanently lower income trajectories.

- **Dominance of child penalty in the gender gap:**
  - As other forms of gender inequality (such as educational attainment and single women's workforce participation) have improved, the penalty from motherhood has become the predominant factor driving gender wage inequality. In Denmark, approximately 80% of the current gender pay gap is attributed to the impact of children, reframing the gender wage gap as primarily a "childbearing pay gap."

- **Intergenerational transmission of norms:**
  - The study uncovers intergenerational effects, indicating daughters whose mothers experienced significant career setbacks due to childbearing are more likely to encounter similar penalties themselves. This suggests cultural or behavioral norms around family roles and career expectations are transmitted across generations, indicating that addressing these deep-seated norms requires sustained policy efforts.

- **Graphical Results:**
  - Visual representations show dramatic earnings trajectories for women post-childbirth, highlighting the sharp and persistent divergence from men's stable earnings paths. Graphs clearly demonstrate declines in women's employment rates, hours worked, and hourly wage rates immediately following childbirth, none of which rebound fully over subsequent years. These visualizations effectively illustrate the severe and lasting economic consequences of motherhood on women's careers.

#### Methodological Notes

The study's event-study approach, supported by robust administrative data, offers compelling causal evidence linking childbirth directly to women's earnings penalties. By comparing pre- and post-childbirth earnings trajectories and using men as a control group, the study isolates the childbirth effect convincingly. The Danish context provides additional credibility, as the persistence of significant penalties in a country with strong gender-equality policies suggests even greater challenges elsewhere. The findings underline the necessity of supportive policies such as affordable childcare, paternal leave, and flexible working conditions to mitigate the motherhood penalty effectively.

### 5. Intersection of Gender, Race, and Region in Wage Gaps (Chapman & Benis 2017)

#### Research Scope & Methods
Chapman and Benis examine the geographical dimensions of the gender wage gap in the U.S., analyzing how regional variations affect women’s earnings beyond individual characteristics.  
Using data from the National Women’s Law Center, they perform difference-in-means tests and linear regression models to evaluate the influence of regional location.  
The study controls for state-level economic and demographic factors to isolate the independent effects of geography.

#### Key Findings
- **Significant regional variation**:  
  Gender wage gaps vary widely across U.S. states. Northeastern states generally have smaller gaps, while Southern and Midwestern states show larger disparities. Latina women experience particularly pronounced disparities.

- **Independent regional effects**:  
  Even after controlling for economic conditions and demographics, regional location significantly impacts wage disparities, suggesting that local policies, norms, and discrimination patterns matter.

- **Intersectional insights**:  
  The magnitude of wage gaps for Latina and Black women varies greatly by region, highlighting the intersection of gender, race, and geography.

- **Graphical Results**:  
  Visualizations, including U.S. maps and comparative charts, illustrate regional disparities and intersectional patterns in gender wage gaps.

#### Methodological Notes
- The study demonstrates the necessity of incorporating geographical context into gender wage gap analyses.
- Aggregating state-level data with rigorous controls shows that regional-specific policies and cultural norms critically shape wage outcomes.
- Findings suggest the need for **targeted regional and local policy interventions** to address persistent disparities.

### 6. The Other Side of Organized Labor

#### Research Scope & Methods
The paper The Other Side of Organized Labor: Corruption and the Enforceability of Noncompete Agreements by Benmelech, Bergman, and Enrique (2023) examines an underexplored aspect of labor unions—their role in facilitating corruption by influencing the enforcement of noncompete agreements (NCAs). While existing literature has extensively studied unions' effects on wages and worker protections, this paper shifts focus to their potential negative externalities, particularly in contexts where unions collude with employers to restrict worker mobility.

#### Key Findings
- **Unions and Worker Protections**:  
  -Traditional labor economics emphasizes unions' role in improving wages and working conditions (Freeman & Medoff, 1984). However, recent studies suggest that unions may also engage in rent-seeking behavior (Hirsch, 2008). Benmelech et al. contribute to this debate by showing that unions can act as enforcers of NCAs, particularly in corrupt environments where legal institutions are weak. Their empirical analysis reveals that states with higher levels of union corruption exhibit significantly stricter enforcement of NCAs (Figure 2), supporting the hypothesis that unions may collude with firms to suppress competition.

- **Noncompete Agreements and Labor Market Rigidity:**
  -NCAs are typically justified as tools to protect trade secrets (Marx et al., 2009), but they can also stifle innovation and wage growth (Starr et al., 2021). Benmelech et al. extend this literature by demonstrating that unions, rather than always advocating for worker freedom, may help enforce NCAs when doing so benefits entrenched interests. Their regression results (Table 3) show a positive correlation between unionization rates and NCA enforceability in corrupt states, suggesting that union power does not uniformly benefit workers.

  -The paper also intersects with research on corruption and labor markets (Shleifer & Vishny, 1993). This supports theories that institutional capture can distort labor markets (Acemoglu & Robinson, 2019), but uniquely implicates unions as active participants in such capture.

Benmelech et al. provide a counterintuitive perspective on unions, challenging the assumption that they always act in workers' best interests. Their event-study analysis (Figure 5) further shows that corruption scandals lead to a decline in NCA enforceability, reinforcing the link between union malfeasance and labor market restrictions. Future research could explore whether similar dynamics exist in other labor regulations, such as occupational licensing or non-poaching agreements.

### 3.5 Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import dask.dataframe as dd
import statsmodels.formula.api as smf
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the data
df = pd.read_csv('usa_00005.csv.gz')
```

```python
# Get a list of all columns in the DataFrame
existing_columns = df.columns.tolist()

# Columns to drop
columns_to_drop = ['WKSWORK1', 'RACED', 'EDUCD', 'EMPSTATD']

# Find common columns (columns that exist and need to be dropped)
common_columns = list(set(existing_columns) & set(columns_to_drop))

# Drop only the common columns if they exist
if common_columns:
  df = df.drop(columns=common_columns, axis=1)
```

```python
# Individual total income and family total income both are in 10k unit
df['INCTOT'] = df['INCTOT'] / 10000
df['FTOTINC'] = df['FTOTINC'] / 10000
df = df[~df['INCTOT'].isin([0, 9999999, 9999998])]
df = df[df['EMPSTAT'].isin([1, 2])] # 1: Employed, 2: Unemployed
df['INC_SHARE'] = df['INCTOT'] / df['FTOTINC'] *100 # Women's Individual-to-Household Income Ratio
print(df.shape)
print(df['NCHILD'].value_counts())
```

### 3.6 Charts and Plots

Before conducting our in-depth analysis, we performed a preliminary inspection of the dataset. During this initial exploration, we discovered several interesting patterns and observations worth noting.


```python
filtered_df = df[(df['INCTOT'] < df['INCTOT'].quantile(0.99)) & (df['INCTOT'].notna())] # Excluded the extreme outliers
def group_education(code):
    if code in [0, 1, 2, 3, 4, 5]:
        return 'Below High School'
    elif code == 6:
        return 'High School'
    elif code in [7, 8, 9, 10]:
        return 'College'
    elif code == 11:
        return 'Graduate (5+ years)'
    else:
        return 'Other/Unknown'

df['EDUC_GROUP'] = df['EDUC'].apply(group_education)
female_df = df[df['SEX'] == 2]
male_df = df[df['SEX'] == 1]
female_educ_counts = female_df['EDUC_GROUP'].value_counts().sort_index()
male_educ_counts = male_df['EDUC_GROUP'].value_counts().sort_index()


emp_df = df[df['EMPSTAT'].isin([1, 2]) & df['SEX'].isin([1, 2])]
emp_counts = emp_df.groupby(['SEX', 'EMPSTAT']).size().unstack().fillna(0)
emp_counts.index = emp_counts.index.map({1: 'Male', 2: 'Female'})
emp_counts.columns = ['Employed', 'Unemployed']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.boxplot(
    data=filtered_df,
    x='SEX',
    y='INCTOT',
    palette={1: 'lightblue', 2: 'lightpink'}, 
    ax=axes[0, 0]
)
axes[0, 0].set_xticks([0, 1])
axes[0, 0].set_xticklabels(['Male', 'Female'])
axes[0, 0].set_title('Income Comparison by Gender')
axes[0, 0].set_xlabel('Gender')
axes[0, 0].set_ylabel('Income (10k USD)')

emp_counts.plot(kind='bar', ax=axes[0, 1], color=['skyblue', 'salmon'], legend=True)
axes[0, 1].set_title('Employment Status by Gender')
axes[0, 1].set_xlabel('Gender')
axes[0, 1].set_ylabel('Count')
axes[0, 1].legend(title='Employment Status')
axes[0, 1].tick_params(axis='x', rotation=0)

axes[1, 0].pie(female_educ_counts, labels=female_educ_counts.index,
               autopct='%1.1f%%', startangle=90, explode=[0.05]*len(female_educ_counts),
               shadow=True)
axes[1, 0].set_title('Education Distribution (Female)')
axes[1, 0].axis('equal')

axes[1, 1].pie(male_educ_counts, labels=male_educ_counts.index,
               autopct='%1.1f%%', startangle=90, explode=[0.05]*len(male_educ_counts),
               shadow=True)
axes[1, 1].set_title('Education Distribution (Male)')
axes[1, 1].axis('equal')

plt.tight_layout()
plt.show()
```

### 4.1. Preliminary Regression analysis on Relevant Factors

Before conducting a deeper analysis on our topic of interest, we first aim to identify which variables in our dataset are potentially influential in determining women's income. To achieve this, we perform a series of preliminary simple regressions. This initial step serves two purposes:

1. To eliminate variables that show only weak correlation with women's income.  
2. To highlight and retain the most relevant factors for further multivariate analysis.

This filtering process helps streamline our subsequent modeling and ensures we focus on the variables that meaningfully contribute to income variation among women.



### a. Race

Rather than applying a linear regression model to the `RACE` variable, we chose to treat it as a categorical variable with multiple levels. This is because `RACE` is encoded as discrete categories ranging from 1 to 9, each representing a different racial group, without any inherent numerical ordering.

Using regression directly on `RACE` as a numeric predictor would incorrectly assume a linear relationship between race codes and income, which is not meaningful. Instead, we opted for visualization techniques—such as boxplots and bar plots—to examine the distribution and average income across different racial groups. These methods allow us to capture income disparities across categories without imposing inappropriate numerical assumptions.



```python
df = df[df['SEX'] == 2] # For research purpose, we only keep females
# regression on race
race_mapping = {
    1: 'White',
    2: 'Black/African American',
    3: 'American Indian',
    4: 'Chinese',
    5: 'Japanese',
    6: 'Other Asian or Pacific Islander',
    7: 'Other race, nec',
    8: 'Two major races',
    9: 'Three or more major races'
}
df['RACE_LABEL'] = df['RACE'].map(race_mapping)
filtered_df = df[df['INCTOT'] < df['INCTOT'].quantile(0.99)].copy()
filtered_df['INCTOT'] = filtered_df['INCTOT']

race_income = filtered_df.groupby('RACE_LABEL')['INCTOT'].mean().reset_index()
race_income = race_income.sort_values(by='INCTOT', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=race_income, x='RACE_LABEL', y='INCTOT', palette='Blues_d')
plt.title('Average Income by Race')
plt.xlabel('Race')
plt.ylabel('Average Income (10K USD)')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

The first figure presents a simple bar plot comparing the **average income** across racial groups.

We then apply a **categorical OLS regression**, treating `RACE` as a factor variable. This approach allows us to:
- Quantify how much each racial group’s income differs from a chosen **reference group** (in this case, White individuals),
- Assess whether those differences are **statistically significant**, and
- Construct **confidence intervals** for the estimated differences.

```python
from stargazer.stargazer import Stargazer
from IPython.core.display import HTML
pd.options.display.float_format = '{:.3f}'.format
model = smf.ols('INCTOT ~ C(RACE)', data=df).fit()

stargazer = Stargazer([model])
HTML(stargazer.render_html())

# Customizations
stargazer.title('Regression Results')
stargazer.covariate_order(['Intercept', 'C(RACE)[T.2]', 'C(RACE)[T.3]', 'C(RACE)[T.4]', 'C(RACE)[T.5]', 'C(RACE)[T.6]', 'C(RACE)[T.7]', 'C(RACE)[T.8]', 'C(RACE)[T.9]'])  # Specify the order of variables
stargazer.rename_covariates({'C(RACE)[T.2]': 'Black/African American', 'C(RACE)[T.3]': 'American Indian', 'C(RACE)[T.4]': 'Chinese', 'C(RACE)[T.5]': 'Japanese', 'C(RACE)[T.6]': 'Other Asian or Pacific Islander', 'C(RACE)[T.7]': 'Other race, nec', 'C(RACE)[T.8]': 'Two major races', 'C(RACE)[T.9]': 'Three or more major races'})  # Rename variables
stargazer.add_custom_notes(['Standard errors in parentheses'])

# Print the table as HTML
HTML(stargazer.render_html())
```

The estimated regression equation is:


$$
\begin{align*}
\text{INCTOT} &=\ 5.182 \\
&\quad - 1.003 \cdot D_{\text{Black/African American}} \\
&\quad - 1.513 \cdot D_{\text{American Indian}} \\
&\quad + 2.029 \cdot D_{\text{Chinese}} \\
&\quad + 1.555 \cdot D_{\text{Japanese}} \\
&\quad + 0.928 \cdot D_{\text{Other Asian or Pacific Islander}} \\
&\quad - 1.815 \cdot D_{\text{Other race, nec}} \\
&\quad - 0.614 \cdot D_{\text{Two major races}} \\
&\quad - 0.580 \cdot D_{\text{Three or more major races}} \\
&\quad + \varepsilon
\end{align*}
$$

Where:
- The **baseline group** is *White*.
- Each race is a dummy variable equal to 1 if the individual identifies with that race category, and 0 otherwise.
- $ \varepsilon$ is the error term.


Summary: Race and Income Analysis

The two figures together illustrate how income varies across racial groups, both in raw averages and after adjusting through a categorical regression model.

# 5. Collinearity Analysis

Before proceeding to multivariable regression, it is essential to examine potential collinearity among explanatory variables. Given that variables like age, education, marital status, and childbearing experience are all deeply intertwined in real life, high correlations between them could bias our regression estimates or inflate standard errors.

To ensure the reliability and interpretability of our final model, we perform a collinearity analysis using correlation matrices and Variance Inflation Factors (VIF). This step helps us detect and address any redundancy among predictors, thus improving the robustness and explanatory power of our regression results.



```python
selected_vars = ['AGE', 'NCHILD', 'MARST', 'RACE', 'EDUC']

df_clean = df[selected_vars].dropna()

df_encoded = df_clean.copy()
df_encoded['MARST'] = df_encoded['MARST'].astype('category').cat.codes
df_encoded['RACE'] = df_encoded['RACE'].astype('category').cat.codes
df_encoded['EDUC'] = df_encoded['EDUC'].astype('category').cat.codes

plt.figure(figsize=(10, 8))
sns.heatmap(df_encoded.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Selected Variables', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

X = sm.add_constant(df_encoded)
vif_data = pd.DataFrame()
vif_data['Variable'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif_data = vif_data[vif_data['Variable'] != 'const']
vif_data.sort_values(by='VIF', ascending=False).reset_index(drop=True)
```

**Conclusion:**

The correlation heatmap and VIF table together indicate that there is no serious multicollinearity among the selected explanatory variables (`AGE`, `NCHILD`, `MARST`, `RACE`, `EDUC`). All correlation coefficients are below ±0.4, and all VIF values are well below the commonly accepted threshold of 5. This confirms that these variables can be included in the same regression model without concerns about distorted coefficient estimates due to multicollinearity.

# 7. Multivariate Multiple Regression

### Motivation for Taking the Log of Income

In our multivariate regression model, we use the **natural logarithm of income (`ln(INCTOT)`)** as the dependent variable rather than raw income. This decision is based on several theoretical and empirical considerations:

1. **Income is Highly Skewed**  
Personal income distributions are typically **right-skewed**, with a small number of individuals earning disproportionately high incomes. This skewness violates the **normality assumption** underlying OLS regression, which can distort coefficient estimates and significance tests. Log-transforming income makes the distribution more symmetric and closer to normal.

2. **Heteroskedasticity Concerns**  
Without transformation, the variance of income tends to increase with the level of income, violating the OLS assumption of **constant error variance**. Applying the natural log stabilizes the variance across observations, helping to mitigate **heteroskedasticity** and improving the efficiency of our estimates.

3. **Interpretability of Coefficients**  
In a log-linear model, coefficients can be interpreted approximately as **percentage changes** in income. This provides a more intuitive and meaningful way to understand the relationship between predictors (such as race, education, or marital status) and income levels.

### Final Model Structure

$$
\begin{align*}
\ln(\text{INCTOT}_i) =\ & \beta_0
+ \sum_{r=1}^{8} \beta_r \cdot D_{\text{Race}_r}
+ \beta_9 \cdot \text{AGE}_i
+ \beta_{10} \cdot \text{AGE}^2_i
+ \beta_{11} \cdot \text{EDUC}_i \\
&+ \sum_{m=1}^{5} \beta_{11+m} \cdot D_{\text{MaritalStatus}_m}
+ \beta_{17} \cdot \text{NCHILD}_i
+ \beta_{18} \cdot (\text{NCHILD}_i \times \text{AGE}_i)
+ \varepsilon_i
\end{align*}
$$

Where:
- $\ln(\text{INCTOT}_i)$ is the natural log of individual $i$’s total personal income,
- $D_{\text{Race}_r}$ are dummy variables for race (reference: White),
- $\text{AGE}_i$ and $\text{AGE}_i^2$ capture the **nonlinear age-income relationship**,
- $\text{EDUC}_i$ is education level,
- $D_{\text{MaritalStatus}}$ is marital status dummies (reference: Married with spouse present),
- $\text{NCHILD}_i$ is the number of children,
- $\text{NCHILD}_i \times \text{AGE}_i$ captures the **interaction between motherhood and age**,
- $\varepsilon_i$ is the error term.

This model allows us to more precisely estimate income drivers by addressing potential **nonlinearities**, **skewness**, and **interaction effects** present in the data.


```python
df['AGE_SQ'] = df['AGE'] ** 2
df['LN_INCTOT'] = np.log(df['INCTOT'])
model_final = smf.ols(
    'LN_INCTOT ~ C(RACE) + AGE + AGE_SQ + EDUC + C(MARST) + NCHILD + NCHILD:AGE',
    data=df
).fit(cov_type='HC3')

stargazer = Stargazer([model_final])
stargazer.title('Final Regression Results')
stargazer.rename_covariates({
    'C(RACE)[T.2]': 'Black',
    'C(RACE)[T.3]': 'American Indian/Alaska Native',
    'C(RACE)[T.4]': 'Chinese',
    'C(RACE)[T.5]': 'Japanese',
    'C(RACE)[T.6]': 'Other Asian/Pacific Islander',
    'C(RACE)[T.7]': 'Other Race',
    'C(RACE)[T.8]': 'Two Major Races',
    'C(RACE)[T.9]': 'Three+ Major Races',
    'AGE': 'Age',
    'AGE_SQ': 'Age Squared',
    'EDUC': 'Level of Education',
    'C(MARST)[T.2]': 'Married, Spouse Absent',
    'C(MARST)[T.3]': 'Separated',
    'C(MARST)[T.4]': 'Divorced',
    'C(MARST)[T.5]': 'Widowed',
    'C(MARST)[T.6]': 'Never Married/Single',
    'NCHILD': 'Number of Children',
    'NCHILD:AGE': 'Children × Age Interaction'
})


HTML(stargazer.render_html())
```

**Summary:**

The regression results show that most predictors significantly influence women's logged income at the 1% level. **Age** has a positive effect initially (coefficient = 0.117) but diminishes as women grow older, as indicated by the negative **AGE²** term (-0.001), confirming a concave (inverted U-shaped) relationship between age and income.

**Marital status** matters: compared to married women with a spouse present, women who are married but spouse absent, separated, or never married earn significantly less, while widowed women earn more. **Race** also shows large disparities: Black and American Indian women earn less than White women, while Asian and Pacific Islander women earn more. **Education** strongly boosts income (coefficient = 0.170), highlighting its key role in earnings potential.

Interestingly, **number of children (NCHILD)** alone raises income slightly (coefficient = 0.175), but the negative interaction with age (-0.005) suggests that as women with more children age, their income advantage diminishes.

Overall, the model explains about **30%** of the variation in income (**R² = 0.297**), and all major theoretical expectations—nonlinearity, race and marital disparities, education benefits, and motherhood penalties—are supported by the data.


# 8. Robustness Test

To ensure the robustness of our findings, we estimate several alternative specifications of the regression model. These allow us to test whether key relationships hold under different modeling choices.



### Test 1: Using Linear Income

To test the robustness of our findings, we re-estimated the model using **raw income (`INCTOT`)** instead of its log. The results remain broadly consistent: key predictors such as **race, age, education, marital status, and number of children** all remain statistically significant, with signs and relative magnitudes similar to those in the log-income model. The negative quadratic term on age confirms a nonlinear income-age relationship, and the **interaction between children and age** remains negative, supporting the idea that the income impact of motherhood declines with age. While the model fit is lower (R² = 0.162), the consistency in direction and significance of the coefficients reinforces the validity of our main results.

# 9. Conclusion

Our analysis demonstrates that motherhood imposes a substantial penalty on women's income in the U.S. labor market. Across all age groups, an increase in the number of children is consistently associated with lower earnings, with the largest negative impact observed among women aged 30–40 — a critical period for both career advancement and family formation. While the effect weakens slightly for older age groups, it remains statistically significant throughout the working life span.

Beyond the number of children, we find that factors such as race, education, marital status, and age strongly influence income levels. Notably, racial disparities persist, with women identifying as Black, American Indian, and other minorities earning significantly less compared to White women. Higher educational attainment and being married with a spouse present are associated with better income outcomes.

Importantly, our robustness checks with heteroskedasticity-consistent standard errors confirm the reliability of our findings, and VIF checks assure that multicollinearity is not a concern in our models.

In sum, the U.S. labor market still exhibits a motherhood penalty, albeit within a broader context of demographic and social factors influencing women's earnings. Recognizing and addressing these disparities remains essential for promoting gender equity and supporting working mothers.

Looking forward, future work could be strengthened by incorporating additional variables such as occupation, industry, and work hours, allowing for better control of selection into different types of employment. Moreover, introducing interaction terms between race and education would provide a deeper understanding of how the motherhood penalty varies across different social groups, offering richer and more nuanced insights.
